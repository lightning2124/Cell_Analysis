# PBMC Detection and Visualization

This Jupyter notebook processes microscope images of Peripheral Blood Mononuclear Cells (PBMCs) 
and uses a YOLOv8 model to detect and visualize the cells.

## Key Components:
1. **Image Processing**: Uses the `process_image` module to preprocess the raw microscope images.
2. **YOLO Model**: Employs a pre-trained YOLOv8 model (`annotator.pt`) for PBMC detection.
3. **Visualization**: Creates a 2x2 subplot for each group of four related images, showing 
   detected PBMCs with bounding boxes.

## Dependencies:
- **os**: For file and directory operations
- **matplotlib**: For creating and displaying plots
- **ultralytics**: For loading and using the YOLOv8 model
- **inference**: Custom module containing the `plot_inference` function
- **process_image**: Custom module for image preprocessing

## Usage:
- Ensure all dependencies are installed and custom modules (`inference`, `process_image`) are in the same directory.
- Place raw microscope images in the `sample_images` folder.
- Run the notebook to process images, detect PBMCs, and visualize results.

## Output:
- Processed images are saved in the `sample_processed_imgs` folder.
- For each group of four related images, a 2x2 subplot is displayed showing:
  * Detected PBMCs with blue bounding boxes
  * The number of detected PBMCs for each image
  * The total number of detected PBMCs across all four images

**Note**: This notebook uses `annotator.pt`, which is a prototype model that my principal investigator has allowed me to demo. The final version is not publicly available and is capable of detecting even more classes of cellular structures.

```python
import os
import matplotlib.pyplot as plt
from ultralytics import YOLO
from inference import plot_inference
import process_image

process_image.process_images("sample_images", "sample_processed_imgs")

folder_path = "sample_processed_imgs"

# Group images by base name
image_groups = {}
for filename in os.listdir(folder_path):
    base_name = os.path.splitext(filename)[0][:-1]  # Extract base name without the suffix
    if base_name not in image_groups:
        image_groups[base_name] = []
    image_groups[base_name].append(os.path.join(folder_path, filename))

model = YOLO('annotator.pt')

# Loop through each group and display the 2x2 subplot
for base_name, group in image_groups.items():
    fig, axes = plt.subplots(2, 2, figsize=(7, 7))
    axes = axes.flatten()

    total_annotations = 0
    annotation_counts = {}

    for i, image_path in enumerate(group):
        # Perform inference once
        result = model(image_path)[0]
        num_annotations = len(result.boxes)
        
        # Use plot_inference function with the existing result
        result_img = plot_inference(image_path, model_path='annotator.pt', existing_result=result)
        
        image_suffix = os.path.splitext(os.path.basename(image_path))[0][-1]
        annotation_counts[image_suffix] = num_annotations
        total_annotations += num_annotations

        # Display on the correct subplot
        axes[i].imshow(result_img)
        axes[i].set_title(f"{image_suffix}: {annotation_counts[image_suffix]}")
        axes[i].axis('off')

    fig.suptitle(f"{base_name} - Total Annotations: {total_annotations}", fontsize=16)
    plt.tight_layout()
    plt.show()
```
